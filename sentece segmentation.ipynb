```python
import requests
from bs4 import BeautifulSoup
url = 'https://en.wikipedia.org/wiki/Dragon_Ball_Z'
response = requests.get(url)
html = response.content

soup = BeautifulSoup(html)
k=soup('p')
text = soup.find('p').get_text()
j=''
for k in text:
    if k=='.':
        j+='.\n';
    else:
        j+=k
print (j)
    
```
